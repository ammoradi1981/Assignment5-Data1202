#### 1a) grading_cat1 - Create this new column using np.select function that we have discussed in the class.

```python
import numpy as np  

# df is DataFrame 

       
conditions = [
    (df['grade'] == 'A++') | (df['grade'] == 'A+') | (df['grade'] == 'A') | (df['grade'] == 'A-'),
    (df['grade'] == 'B+') | (df['grade'] == 'B') | (df['grade'] == 'B-'),
    (df['grade'] == 'C+') | (df['grade'] == 'C'),
    (df['grade'] == 'D') | (df['grade'] == 'D-') 
]

output = ['A','B','C','D']  

# Remove existing grading_cat1 column if it exists
if 'grading_cat1' in df.columns:
    df.drop(columns=['grading_cat1'], inplace=True)

# Add new grading_cat1 column
df['grading_cat1']= np.select(conditions, output, default='TBD')
```

```python
df.head(15)
```

```python
# Remove existing grading_cat1 column if it exists
if 'grading_cat1' in df.columns:
    df.drop(columns=['grading_cat1'], inplace=True)

# Add new 'grading_cat1' column immediately after 'grade' column
df.insert(df.columns.get_loc('grade') + 1, 'grading_cat1', np.select(conditions, output, default='TBD'))
```

#### 1b) grading_cat2 - Create this new column using python functions (def , if else condition)

```python

# Adding 'grading_cat2' column using Python functions  (def , if else condition)

def categ_grade(grade):
    if grade in ['A++', 'A+', 'A', 'A-']:
        return 'A'
    elif grade in ['B+', 'B', 'B-']:
        return 'B'
    elif grade in ['C+', 'C']:
        return 'C'
    elif grade in ['D', 'D-']:
        return 'D'
    else:
        return 'TBD'

# Remove existing grading_cat2 column if it exists
if 'grading_cat2' in df.columns:
    df.drop(columns=['grading_cat2'], inplace=True)


#df['grading_cat2'] = df['grade'].apply(categ_grade)
df.insert(df.columns.get_loc('grade') + 2, 'grading_cat2',df['grade'].apply(categ_grade) )
```

### Pie Chart

```python
import matplotlib.pyplot as plt

# Filter the dataframe to include only rows where 'uploads' is greater than 6000
df_high_upload = df_grouped[df_grouped['uploads'] > 6000]

# Calculate the total uploads for visualization
total_uploads = df_high_upload['uploads'].sum()

# Calculate the percentage of uploads for each country
df_high_upload['upload_percentage'] = (df_high_upload['uploads'] / total_uploads) * 100

# Plotting the pie chart
plt.figure(figsize=(8, 8))
plt.pie(df_high_upload['upload_percentage'], labels=df_high_upload['country'], autopct='%1.1f%%', startangle=140)
plt.title('Uploads by Country with Uploads > 6000')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming 'df' is your DataFrame containing the dataset

# Step 1: Extract relevant columns
country_youtube_data = df[['country', 'videoviews']]

# Step 2: Aggregate data to find total YouTube usage for each country
country_youtube_usage = country_youtube_data.groupby('country')['videoviews'].sum().reset_index()

# Step 3: Select top five countries with highest YouTube usage
top_five_countries = country_youtube_usage.nlargest(5, 'videoviews')

# Step 4: Plot pie chart
plt.figure(figsize=(8, 6))
plt.pie(top_five_countries['videoviews'], labels=top_five_countries['country'], autopct='%1.1f%%', startangle=140)
plt.title('Top Five Countries Using YouTube')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```
